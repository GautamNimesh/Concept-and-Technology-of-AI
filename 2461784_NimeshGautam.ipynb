3.1 Problem - 1A - Single Year HDI Exploration (Latest Year: 2022)

```python
import pandas as pd
import numpy as np
```

```python
df_hdi = pd.read_csv("/content/drive/MyDrive/Concepts and Technologies of AI/Human_Development_Index_Dataset.csv", encoding= "latin1")
```

```python
df_hdi.head(10)
```

1. Extract Latest Year:

```python
print("Unique years in dataset:\n", sorted(df_hdi['year'].unique()))
hdi_2022_df = df_hdi[df_hdi['year'] == 2022].copy()
```

2. Data Exploration:

```python
print(hdi_2022_df.head(10))
```

```python
print("Row, Columns", hdi_2022_df.shape)
print(hdi_2022_df.dtypes)
```

3. Missing Values & Data Cleaning:

```python
numeric_cols = ['hdi', 'gross_inc_percap', 'life_expectancy', 'gender_development']
for col in numeric_cols:
    if col in hdi_2022_df.columns:
        hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

hdi_2022_df.drop_duplicates(inplace=True)
hdi_2022_df.dropna(subset=['hdi'], inplace=True)
```

```python
print("After cleaning, (Row, Column) is ", hdi_2022_df.shape)
```

4. Basic Statistics:

```python
print("Mean HDI:", hdi_2022_df['hdi'].mean())
print("Median HDI:", hdi_2022_df['hdi'].median())
print("Std HDI:", hdi_2022_df['hdi'].std())

print("Highest HDI Country:",
      hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax(), 'country'])
print("Lowest HDI Country:",
      hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(), 'country'])
```

5. Filtering and Sorting:

```python
filtered = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]
hdi_2022_df_sorted = filtered.sort_values(by='gross_inc_percap', ascending=False)
print("Countries with HDI greater than 0.800: \n",hdi_2022_df_sorted[['country', 'hdi', 'gross_inc_percap']].head(10))
```

6. Adding HDI Category Column:

```python
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(hdi_category)
print("\nHDI Category distribution:\n", hdi_2022_df['HDI_Category'].value_counts())
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

3.2 Problem - 1B - HDI Visualization and Trend Analysis (2020 – 2022)

```python
import matplotlib.pyplot as plt
```

1. Data Extraction:

```python
df_hdi_1b = df_hdi[df_hdi['year'].isin([2020, 2021, 2022])].copy()
```

2. Data Cleaning:

```python
df_hdi_1b.replace("–", np.nan, inplace=True)
```

```python
for col in ['hdi', 'gross_inc_percap', 'life_expectancy', 'gender_development']:
    if col in df_hdi_1b.columns:
        df_hdi_1b[col] = pd.to_numeric(df_hdi_1b[col], errors='coerce')

df_hdi_1b.dropna(subset=['hdi', 'country', 'year'], inplace=True)
df_hdi_1b.drop_duplicates(inplace=True)
```

Data Saving:

```python
df_hdi_1b.to_csv("HDI_problem1B.csv", index=False)
```

3. Visualization Tasks:

A. Line Chart — HDI Trend (Country-Level):

```python
#line chart
countries = ['Brazil', 'India', 'Nepal', 'China', 'Japan']
plt.figure(figsize = (10,6))
for c in countries:
    data = df_hdi_1b[df_hdi_1b['country'] == c]
    plt.plot(data['year'], data['hdi'], marker = 'o', label=c)
plt.xlabel("Year")
plt.ylabel("HDI")
plt.legend()
plt.title("HDI Trend (2020–2022)")
plt.show()
```

• B. Generate Visualizations:

Bar Chart: Average HDI by Region (2020–2022)

```python
# Bar Chart – Average HDI by Region
selected = df_hdi_1b[df_hdi_1b['country'].isin(countries)].copy()

# Manually map to regions
region_map = {
    'Nepal': 'South Asia',
    'India': 'South Asia',
    'Brazil': 'Latin America',
    'China': 'East Asia',
    'Japan': 'East Asia'
}

selected['region'] = selected['country'].map(region_map)

# Group by Region × Year and compute mean HDI
region_year = (
    selected .groupby(['region', 'year'])['hdi']
    .mean()
    .reset_index()
    .sort_values(by=['region', 'year'])
)

print(region_year)

# Pivot for grouped bar chart
pivot_df = region_year.pivot(index='region', columns='year', values='hdi')
pivot_df.plot(kind='bar', figsize=(10,6))

plt.xlabel("Region")
plt.ylabel("Average HDI")
plt.title("Average HDI by Region (2020–2022)")
plt.legend(title="Year")
plt.tight_layout()
plt.show()

```

Box Plot: HDI Distribution for 2020, 2021, and 2022

```python
#box plot
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_hdi_1b, x='year', y='hdi')
plt.xlabel("Year")
plt.ylabel("HDI")
plt.title("HDI Distribution (2020–2022)")
plt.suptitle("")
plt.show()
```

Scatter Plot: HDI vs. GNI per Capita

```python
#Scatter plot
if 'gross_inc_percap' in df_hdi_1b.columns:
    plt.figure()
    sns.regplot(data=df_hdi_1b, x='gross_inc_percap', y='hdi',
            scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.title("HDI vs GNI per Capita")
    plt.show()
else:
    print("GNI per Capita variable not available in the dataset.")
```

#4 Problem 2
Advanced HDI Exploration

1. Create South Asia Subset:

```python
south_asia = ["Afghanistan", "Bangladesh", "Bhutan", "India",
              "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
```

```python
south_asia_df = df_hdi[df_hdi['country'].isin(south_asia)].copy()
```

```python
south_asia_df.to_csv("HDI_SouthAsia.csv", index=False)
```

2. Composite Development Score:

```python
south_asia_df['Composite_Score'] = 0.30 * south_asia_df['life_expectancy'] + 0.30 * south_asia_df['gross_inc_percap']
ranked = south_asia_df.groupby('country')[['hdi', 'Composite_Score']].mean()
top_5 = ranked.sort_values(by='Composite_Score', ascending=False).head(5).reset_index()
print(top_5)

plt.figure()
plt.barh(top_5['country'], top_5['Composite_Score'])
plt.xlabel("Composite Score")
plt.title("Top 5 South Asian Countries")
plt.show()
```

3. Outlier Detection:

```python
def outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    return (series < q1 - 1.5 * iqr) | (series > q3 + 1.5 * iqr)

o = outliers(south_asia_df['hdi']) | outliers(south_asia_df['gross_inc_percap'])
plt.figure()
plt.scatter(south_asia_df['gross_inc_percap'], south_asia_df['hdi'])
plt.scatter(south_asia_df[o]['gross_inc_percap'], south_asia_df[o]['hdi'])
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outlier Detection")
plt.show()
```

4. Exploring Metric Relationships:

```python
print("Pearson correlation of gender_development with HDI: ",
      south_asia_df['gender_development'].corr(south_asia_df['hdi']))
print("Pearson correlation of life_expectancy with HDI: ",
      south_asia_df['life_expectancy'].corr(south_asia_df['hdi']))
```

```python
sns.regplot(data=south_asia_df, x='life_expectancy', y='hdi')
plt.title("Relationship: Life Expectancy vs HDI")
plt.show()

sns.regplot(data=south_asia_df, x='gender_development', y='hdi')
plt.title("Relationship: Gender Development vs HDI")
plt.show()
```

5. Gap Analysis:

```python
south_asia_df['GNI_HDI_Gap'] = south_asia_df['gross_inc_percap'] - south_asia_df['hdi']
descending_gap = south_asia_df.groupby('country')['GNI_HDI_Gap'].mean().head(3).sort_values(ascending=False)
print("\nAscending (Positive Gaps Top 3):")
print(descending_gap)
ascending_gap = south_asia_df.groupby('country')['GNI_HDI_Gap'].mean().head(3).sort_values(ascending=True)
print("\nAscending (Negative Gaps Top 3):")
print(ascending_gap)
```

```python
top3_positive = descending_gap.head(3)
top3_negative = ascending_gap.head(3)

fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Positive Gaps
axes[0].bar(top3_positive.index, top3_positive.values)
axes[0].set_title("Top 3 Positive GNI-HDI Gaps")
axes[0].set_ylabel("Gap Value")

# Negative Gaps
axes[1].bar(top3_negative.index, top3_negative.values)
axes[1].set_title("Top 3 Negative GNI-HDI Gaps")
axes[1].set_ylabel("Gap Value")

plt.tight_layout()
plt.show()
```

#5 Problem 3
Comparative Regional Analysis: South Asia vs Middle East

1. Create Middle East Subset:

```python
middle_east = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
               "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria",
               "United Arab Emirates", "Yemen"]
```

```python
middle_east_df = df_hdi[df_hdi['country'].isin(middle_east)].copy()
```

```python
south_asia_df.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
middle_east_df.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

2. Descriptive Statistics:

```python
print("South Asia Mean HDI:", south_asia_df['hdi'].mean())
print("South Asia Std HDI:", south_asia_df['hdi'].std())

print("Middle East Mean HDI:", middle_east_df['hdi'].mean())
print("Middle East Std HDI:", middle_east_df['hdi'].std())
```

3. Top and Bottom Performers:

```python
sa_mean = south_asia_df.groupby('country')['hdi'].mean().reset_index()
me_mean = middle_east_df.groupby('country')['hdi'].mean().reset_index()
```

```python
south_asia_top = sa_mean.nlargest(3, 'hdi')
south_asia_bottom = sa_mean.nsmallest(3, 'hdi')

middle_east_top = me_mean.nlargest(3, 'hdi')
middle_east_bottom = me_mean.nsmallest(3, 'hdi')
```

```python
#south asia
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].bar(south_asia_top['country'], south_asia_top['hdi'])
axes[0].set_title("South Asia Top 3 HDI")
axes[0].set_xlabel("Country"); axes[0].set_ylabel("HDI")

axes[1].bar(south_asia_bottom['country'], south_asia_bottom['hdi'])
axes[1].set_title("South Asia Bottom 3 HDI")
axes[1].set_xlabel("Country")
plt.tight_layout()
plt.show()
```

```python
#middle east
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].bar(middle_east_top['country'], middle_east_top['hdi'])
axes[0].set_title("Middle East Top 3 HDI")
axes[0].set_xlabel("Country"); axes[0].set_ylabel("HDI")

axes[1].bar(middle_east_bottom['country'], middle_east_bottom['hdi'])
axes[1].set_title("Middle East Bottom 3 HDI")
axes[1].set_xlabel("Country")
plt.tight_layout()
plt.show()
```

4. Metric Comparisons:

```python
metrics = ['gender_development', 'life_expectancy', 'gross_inc_percap']
years = [2020, 2021, 2022]

for metric in metrics:
    sa_avg = south_asia_df.groupby('year')[metric].mean().reindex(years)
    me_avg = middle_east_df.groupby('year')[metric].mean().reindex(years)

    plt.figure()
    plt.bar([a - 0.2 for a in years], sa_avg, width=0.4, label='South Asia')
    plt.bar([a + 0.2 for a in years], me_avg, width=0.4, label='Middle East')
    plt.xlabel("Year")
    plt.ylabel(metric)
    plt.title(metric + " Comparison: South Asia vs Middle East")
    plt.legend()
    plt.show()
```

5. HDI Disparity:

```python
south_asia_range = south_asia_df['hdi'].max() - south_asia_df['hdi'].min()
middle_east_range = middle_east_df['hdi'].max() - middle_east_df['hdi'].min()

south_asia_cv = south_asia_df['hdi'].std() / south_asia_df['hdi'].mean()
middle_east_cv = middle_east_df['hdi'].std() / middle_east_df['hdi'].mean()

print("South Asia HDI Range:", south_asia_range)
print("Middle East HDI Range:", middle_east_range)
print("South Asia HDI CV:", south_asia_cv)
print("Middle East HDI CV:", middle_east_cv)
```

6. Correlation Analysis:

```python
for region_df, region_name in [(south_asia_df, "South Asia"), (middle_east_df, "Middle East")]:
    print(region_name, "Correlation with HDI (Gender Development):",
          region_df['gender_development'].corr(region_df['hdi']))
    print(region_name, "Correlation with HDI (Life Expectancy):",
          region_df['life_expectancy'].corr(region_df['hdi']))

    # Scatter 1
    plt.figure()
    plt.scatter(region_df['gender_development'], region_df['hdi'])
    plt.xlabel("Gender Development")
    plt.ylabel("HDI")
    plt.title(region_name + ": HDI vs Gender Development")
    plt.show()
    # Scatter 2
    plt.figure()
    plt.scatter(region_df['life_expectancy'], region_df['hdi'])
    plt.xlabel("Life Expectancy")
    plt.ylabel("HDI")
    plt.title(region_name + ": HDI vs Life Expectancy")
    plt.show()
```

7. Outlier Detection:

```python
def detect_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    return (series < q1 - 1.5 * iqr) | (series > q3 + 1.5 * iqr)

for region_df, region_name in [(south_asia_df, "South Asia"), (middle_east_df, "Middle East")]:

    agg = region_df.groupby('country')[['hdi', 'gross_inc_percap']].mean()
    hdi_out = detect_outliers(agg['hdi'])
    gni_out = detect_outliers(agg['gross_inc_percap'])
    outliers = agg[hdi_out | gni_out]
    print("\n", region_name, "Outliers:")
    print(outliers)
    plt.figure()
    plt.scatter(agg['gross_inc_percap'], agg['hdi'], label='Normal', alpha=0.7)

    if not outliers.empty:
        plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], label='Outliers', s=120, marker='x')

    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.title(region_name + ": Outlier Detection (HDI vs GNI)")
    plt.legend()
    plt.show()
```
